```python
import pandas as pd
import numpy as np
```

```python
state_data = pd.read_csv("state_longterm_care_list.csv",index_col=0)
state_data['provider_name'] = state_data['Provider Name'].copy().str.lower().str.replace('.', '').str.replace('&', 'and')
state_data.head()
```

```python
federal_data = pd.read_csv("federal_nursing_home_cms.csv",index_col=0)
federal_data['provider_name'] = federal_data['provider_name'].str.lower().str.replace('.', '').str.replace('&', 'and')
federal_data.head()
```

```python
provider_names_unique_to_federal = set(federal_data['provider_name']) - set(state_data['Provider Name'])
provider_names_unique_to_federal = [str(name) for name in provider_names_unique_to_federal]
print(sorted(provider_names_unique_to_federal))
```

```python
provider_names_unique_to_state = set(state_data['provider_name']) - set(federal_data['provider_name'])
print(sorted(provider_names_unique_to_state))
```
